```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os

# Set the style for seaborn
sns.set(style="whitegrid")
# Define the directory where the images are stored
image_dir = os.path.join(os.getcwd(), 'Figures')
print(f"Image directory: {image_dir}")
# Ensure the directory exists
if not os.path.exists(image_dir):
    os.makedirs(image_dir)
```

```python
# Define the range of G values within the interesting interval
G_values = np.linspace(0.5, 2.5, 500)

# Define the equilibrium probabilities for public good provision
def alpha_wta(G):
    # WTA case: 0 if G < 1, 0.5 if 1 <= G < 2, 1 if G >= 2
    return np.piecewise(G, [G < 1, (G >= 1) & (G < 2), G >= 2], [0, 0.5, 1])

def beta_pr(G):
    # PR case: 0 if G < 1, G - 1 if 1 <= G < 2, 1 if G >= 2
    return np.piecewise(G, [G < 1, (G >= 1) & (G < 2), G >= 2], [0, lambda G: G - 1, 1])

# Compute the probabilities
alpha_vals = alpha_wta(G_values)
beta_vals = beta_pr(G_values)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(G_values, alpha_vals, label="WTA: $\\alpha(G)$", linestyle='--')
plt.plot(G_values, beta_vals, label="PR: $\\beta(G)$", linestyle='-')
plt.axvline(1, color='grey', linestyle=':', linewidth=0.8)
plt.axvline(2, color='grey', linestyle=':', linewidth=0.8)

plt.title("Equilibrium probability of public good provision under WTA and PR", fontsize=18)
plt.xlabel("Value of public good ($G$)", fontsize=14)
plt.ylabel("Probability of offering the public good", fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
# Save the figure
plt.savefig(os.path.join(image_dir, 'lp_public_good_provision_probabilities.png'), dpi=300, bbox_inches='tight')
plt.show()
```
