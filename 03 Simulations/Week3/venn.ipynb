```python
# import relevant libraries
import os
from pathlib import Path

import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Set path to the Week 3 figures directory.
# This works whether the notebook is run from the repo root or from this folder.
wd = Path.cwd()
week3_dir = wd / "03 Simulations" / "Week3" if (wd / "03 Simulations" / "Week3").exists() else wd
img_dir = week3_dir / "Figures"
img_dir.mkdir(parents=True, exist_ok=True)
```

```python
# Define labels for the three sets
labels = {
    '100': 'Broad individual/\ngroup-based transfers',
    '010': 'Broad place-based\ncorruption',
    '001': 'Targeted individual/group\n-based corruption',
    '110': 'Broad place-based\npolicies (e.g. large-scale IP)',
    '101': 'Clientelism &\ntargeted ind./\ngroup-based benefits',
    '011': 'Regional\ncorruption and\npork-barrel\npolitics',
    '111': 'Pork-barrel politics',
}

# Create the Venn diagram
plt.figure(figsize=(10, 8))
venn = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels=('Legal', 'Geographic', 'Selective'))

# Annotate the regions manually
for region_code, label in labels.items():
    subset = venn.get_label_by_id(region_code)
    if subset:
        subset.set_text(label)
        subset.set_fontsize(8)
plt.savefig(os.path.join(img_dir, 'venn_diagram_v1.png'), dpi=300, bbox_inches='tight')
# plt.title("Conceptual Venn Diagram of Distributive Politics")
plt.show()
```

```python
# Three binary dimensions of distributive politics:
#   Unit: non-territorial vs territorial
#   Scope: broad vs narrow
#   Mode: rule-bound vs discretionary
#
# The Venn circles represent the marked categories: territorial, narrow, and discretionary.
# The outside-all-circles region is therefore non-territorial, broad, and rule-bound.
labels = {
    "100": "Place-based\npolicy",
    "010": "Targeted\nsocial policy",
    "001": "Politicised\nmass transfers",
    "110": "Targeted local\npublic goods",
    "101": "Pork-barrel /\nregional patronage",
    "011": "Patronage,\nclientelism\n& selective\nsubsidies",
    "111": "Group-based\nlocal clientelism",
}

fig, ax = plt.subplots(figsize=(10, 8))
venn = venn3(
    subsets=(1, 1, 1, 1, 1, 1, 1),
    set_labels=("Territorial", "Narrow", "Discretionary"),
    ax=ax,
)

for region_code, label in labels.items():
    subset = venn.get_label_by_id(region_code)
    if subset:
        subset.set_text(label)
        subset.set_fontsize(7)

for set_label in venn.set_labels:
    if set_label:
        set_label.set_fontsize(15)

# Label the unmarked baseline category outside all three circles.
# ax.text(
#     -0.78,
#     0.58,
#     "Universalistic\nwelfare provision\n(non-territorial, broad,\nrule-bound)",
#     ha="center",
#     va="center",
#     fontsize=8,
#     bbox={"boxstyle": "round,pad=0.35", "facecolor": "white", "edgecolor": "0.75", "alpha": 0.9},
# )

#ax.set_title("Unit, scope, and mode of allocation", fontsize=16, pad=18)
plt.savefig(img_dir / "venn_diagram_v2.png", dpi=300, bbox_inches="tight")
plt.show()
```
