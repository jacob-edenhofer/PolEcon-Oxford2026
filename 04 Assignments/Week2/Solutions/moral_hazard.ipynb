# Week 2, §4.1 — Moral hazard in electoral teams

$N$ candidates choose hidden effort $e_i$ at cost $c(e_i) = e_i^2/2$. Team
output is $f(\sum_j e_j) = \alpha \sum_j e_j$. We compare equal sharing
($\phi \equiv 1/N$) with rank-based rewards ($\phi(r) = \beta/r$).


```python
import numpy as np
import matplotlib.pyplot as plt
```

## (a) Equal sharing

Symmetric Nash effort: $e^* = \alpha/N$. Total output: $\alpha$ (constant
in $N$). Planner's first-best: $e^{FB} = \alpha$, total $N\alpha$. The
wedge $e^*/e^{FB} = 1/N$ widens as $N \to \infty$ — classic free-riding.


```python
alpha = 1.0
Ns = np.array([2, 5, 10, 50])
e_equal = alpha / Ns
total_equal = Ns * e_equal  # = alpha for all N
for N, e, T in zip(Ns, e_equal, total_equal):
    print(f'N = {N:3d}: e_i* = {e:.4f}, total = {T:.4f}')
```

## (b) Rank-based rewards $\phi(r) = \beta/r$

FOC: $e_i^* = \alpha\beta/r_i$. Higher rank (smaller $r_i$) implies higher
effort. Total effort: $\alpha\beta H_N$ where $H_N = \sum_{r=1}^N 1/r$ is
the harmonic number.


```python
alpha, beta = 1.0, 1.0
for N in Ns:
    ranks = np.arange(1, N + 1)
    e_rank = alpha * beta / ranks
    print(f'N = {N:3d}: top effort = {e_rank[0]:.4f}, '
          f'bottom effort = {e_rank[-1]:.4f}, total = {e_rank.sum():.4f}')
```

## (c) Comparison: total team effort vs.\ $N$


```python
Ns_full = np.arange(2, 100)
totals_equal = np.full_like(Ns_full, alpha, dtype=float)
totals_rank = np.array([
    sum(alpha * beta / r for r in range(1, N + 1)) for N in Ns_full
])
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(Ns_full, totals_equal, 'C0-', label='equal sharing')
ax.plot(Ns_full, totals_rank, 'C1-', label=r'rank-based ($\phi(r) = 1/r$)')
ax.set(xlabel='$N$', ylabel='total team effort',
       title='Rank-based rewards dominate equal sharing for all $N \\geq 2$')
ax.legend(); ax.grid(alpha=0.3); plt.tight_layout(); plt.show()
```

## (d) Closed vs.\ open list — discussion

* **Closed list** fixes $\phi(r)$ pre-election; rank-based incentives
  operate cleanly. Top of list has the highest $\phi$ and exerts the most
  effort.
* **Open list** lets voters reshuffle; $\phi(r)$ becomes endogenous and
  depends on voter inferences about effort.

Open list elicits **less** effort than closed list when voter signals about
effort are noisy: candidates expect approximately uniform $\phi$ across
ranks, blunting the incentive. The closer voters' signals are to flat noise,
the more open list converges to the equal-sharing benchmark.
